## Importing required libraries

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
import folium # map rendering library
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

## Web scraping for getting required data

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
from IPython.display import display_html
data = str(soup.table)
display_html(data,raw=True)
df = pd.read_html(data)
df = df[0]
df.head()
```

## Data cleaning and manipulation

```python
df = df[df['Borough'] != 'Not assigned']
df = df.groupby(['Postal code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)
df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned',df['Borough'], df['Neighborhood'])
df.head()
```

```python
df.shape
```

## Importing latitude and longtitude values of postal codes 

```python
!wget -q -O 'Geospatial_Coordinates.csv' 
print('Data downloaded!')
```

```python
dfGeo = pd.read_csv('https://cocl.us/Geospatial_data')
dfGeo.head()
```

## Combining latitude and longtitude values with our data frame

```python
df = df.merge(dfGeo, left_on = 'Postal code', right_on = 'Postal Code')
df = df[['Postal code', 'Borough', 'Neighborhood', 'Latitude', 'Longitude']]
df.rename(columns={'Postal code':'PostalCode'}, inplace = True)
df.head()
```

## Visualize & Cluster Neighborhoods

```python
#Getting all the rows from the data frame which contains Toronto in their Borough.
dfToronto = df[df['Borough'].str.contains('Toronto')]
dfToronto.head()
```

```python
dfToronto.shape
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(dfToronto['Latitude'],dfToronto['Longitude'],dfToronto['Borough'],dfToronto['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# set number of clusters
kclusters = 5

toronto_clustering = dfToronto.drop(['PostalCode','Borough','Neighborhood'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
dfToronto.insert(0, 'Cluster Labels', kmeans.labels_)
dfToronto.head()
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color schema for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(dfToronto['Latitude'], dfToronto['Longitude'], dfToronto['Neighborhood'], dfToronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
